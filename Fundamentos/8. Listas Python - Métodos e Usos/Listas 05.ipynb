# Algumas Funções Básicas de Lista

## Tamanho da Lista

tamanho = len(lista)

```python
produtos = ['apple tv', 'mac', 'iphone x', 'IPad', 'apple watch', 'mac book', 'airpods', 'airpods 4']
```

- Quantos produtos temos a venda?

```python
tamanho = len(produtos)
print(f'Temos {tamanho} produtos.')
```

## Maior e Menor Valor

maior = max(lista)

menor = min(lista)

```python
vendas = [1000, 1500, 15000, 270, 900, 100, 1200]
```

- Qual o item mais vendido?
- Qual o item menos vendido?

```python
maximo = max(vendas)
minimo = min(vendas)
print(f'O máximo de itens vendidos foi {maximo} e o mínimo foi {minimo}.')
```

```python
produto_mais_vendido = vendas.index(maximo)
produto_menos_vendido = vendas.index(minimo)

print(f'O produto MAIS vendido foi {produtos[produto_mais_vendido]} e o MENOS vendido foi {produtos[produto_menos_vendido]} ')
```
