#### 1. Faça um programa que peça uma nota, entre zero e dez. Mostre uma mensagem caso o valor seja inválido e continue pedindo até que o usuário informe um valor válido.

````python
nota = int(input('Digite uma nota entre 0 e 10: '))

while nota > 10 or nota < 0:
    nota = int(input('Valor inválido, digite uma nota entre 0 e 10: '))

print(f'A nota {nota} é VÁLIDA.')
````

#### 2. Faça um programa que leia um nome de usuário e a sua senha e não aceite a senha igual ao nome do usuário, mostrando uma mensagem de erro e voltando a pedir as informações.

````python
login = input('Digite o nome do usuário: ')
senha = input('Digite a senha: ')

while login == senha:
    login = input('ERRO: LOGIN E SENHA IGUAIS! Digite o nome do usuário: ')
    senha = input('ERRO: LOGIN E SENHA IGUAIS! Digite a senha: ')
print(login, senha)
    
````

#### 3. Faça um programa que leia e valide as seguintes informações (e para cada uma delas, continue pedindo a informação até o usuário inserir corretamente):
##### Nome: maior que 3 caracteres;
##### Idade: entre 0 e 150;
##### Salário: maior que zero;
##### Sexo: 'f' ou 'm';
##### Estado Civil: 's', 'c', 'v', 'd';

````python
nome = input('Digite o seu nome: ')
while len(nome) < 4:
    nome = input('Nome inválido! Digite um nome com mais de 3 caracteres: ')

idade = int(input('Digite a idade: '))
while idade < 0 or idade > 150:
    idade = int(input('Idade inválida! Digite uma idade entre 0 e 150: '))

salario = float(input('Digite o salário: '))
while salario <= 0:
    salario = float(input('Salário inválido! Digite um salário maior que zero: '))

sexo = input('Digite o sexo (m/f): ')
while sexo != 'f' and sexo != 'm':
    sexo = input('Sexo inválido! Digite "m" para masculino ou "f" para feminino: ')

estadocivil = input('Estado Civil (s/c/v/d): ')
while estadocivil != 's' and estadocivil != 'c' and estadocivil != 'v' and estadocivil != 'd':
    estadocivil = input('Inválido! Use "s", "c", "v" ou "d": ')

print(nome, idade, salario, sexo, estadocivil)
````

#### 5. Altere o programa anterior permitindo ao usuário informar as populações e as taxas de crescimento iniciais. Valide a entrada e permita repetir a operação.

````python
while True:
    a_habitantes = float(input('Digite o número de habitantes da cidade A: '))
    while a_habitantes <= 0:
        a_habitantes = float(input('Valor inválido! Digite o número de habitantes da cidade A: '))

    b_habitantes = float(input('Digite o número de habitantes da cidade B: '))
    while b_habitantes <= 0:
        b_habitantes = float(input('Valor inválido! Digite o número de habitantes da cidade B: '))

    a_taxa = float(input('Digite a taxa de crescimento da cidade A (em %): ')) / 100
    while a_taxa <= 0:
        a_taxa = float(input('Taxa inválida! Digite a taxa da cidade A (em %): ')) / 100

    b_taxa = float(input('Digite a taxa de crescimento da cidade B (em %): ')) / 100
    while b_taxa <= 0:
        b_taxa = float(input('Taxa inválida! Digite a taxa da cidade B (em %): ')) / 100

    ano = 0
    while a_habitantes < b_habitantes:
        a_habitantes *= (1 + a_taxa)
        b_habitantes *= (1 + b_taxa)
        ano += 1

    print(f'Em {ano} anos, a cidade A terá mais habitantes que a cidade B.')

    repetir = input('Deseja repetir? (s/n): ')
    if repetir != 's':
        break
````
